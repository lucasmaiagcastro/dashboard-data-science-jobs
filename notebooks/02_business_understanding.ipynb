```python
def parse_skills(value):
    if pd.isna(value):
        return []

    if isinstance(value, list):
        return value

    try:
        parsed = ast.literal_eval(value)

        if isinstance(parsed, list):
            return [
                str(skill).strip().lower()
                for skill in parsed
            ]
    except:
        pass

    return [
        skill.strip().lower()
        for skill in str(value).split(",")
        if skill.strip()
    ]
```

```python
df["skills_list"] = df["skills"].apply(parse_skills)

df[["skills", "skills_list"]].head()
```

```python
df_skills = (
    df.explode("skills_list")
      .rename(columns={"skills_list": "skill"})
)

df_skills = df_skills[df_skills["skill"].notna()]
df_skills = df_skills[df_skills["skill"] != ""]

df_skills.head()
```

```python
top_skills = (
    df_skills["skill"]
    .value_counts()
    .head(20)
)

top_skills
```

```python
skill_frequency = (
    df_skills
    .groupby("skill")["job_title"]
    .count()
    .sort_values(ascending=False)
    .reset_index(name="frequencia")
)

skill_frequency["percentual_vagas"] = (
    skill_frequency["frequencia"] / len(df) * 100
).round(2)

skill_frequency.head(20)
```

```python
def extract_salary_values(value):

    if pd.isna(value):
        return pd.Series([np.nan, np.nan, np.nan])

    numbers = re.findall(
        r"\d[\d,]*",
        str(value)
    )

    numbers = [
        int(num.replace(",", ""))
        for num in numbers
    ]

    if len(numbers) == 0:
        return pd.Series([np.nan, np.nan, np.nan])

    if len(numbers) == 1:
        salary_min = numbers[0]
        salary_max = numbers[0]
    else:
        salary_min = min(numbers)
        salary_max = max(numbers)

    salary_avg = (
        salary_min + salary_max
    ) / 2

    return pd.Series([
        salary_min,
        salary_max,
        salary_avg
    ])
```

```python
df[
    ["salary_min",
     "salary_max",
     "salary_avg"]
] = df["salary"].apply(
    extract_salary_values
)

df[
    [
        "salary",
        "salary_min",
        "salary_max",
        "salary_avg"
    ]
].head(20)
```
